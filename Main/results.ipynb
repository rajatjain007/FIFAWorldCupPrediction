```python
import matplotlib.pyplot as plt
methods = ["Decision Trees","KNN","Logistic Regression","Naive Bayes"]
accuracies = [75.0,31.25,68.75,43.75]
plt.title("DMTs vs Accuracy (Our Implementation)")
plt.xlabel("DMTs")
plt.ylabel("Accuracy(%)")
plt.bar(methods[0], accuracies[0], color = "red", label = "Decision Trees")
plt.bar(methods[1], accuracies[1], color = "lime", label = "KNN")
plt.bar(methods[2], accuracies[2], color = "purple", label = "Logistic Regression")
plt.bar(methods[3], accuracies[3], color = "blue", label = "Naive Bayes")
plt.legend()
plt.show()
```
